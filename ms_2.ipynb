```python
import random
```

на пару в 12:50.
когда должен вставать студент, чтобы в 95 случаях из 100 успеть на пару?

```python
N = 1000
time_in_minutes = []
for i in range(N):
    minutes_exercise = random.randint(8,12)
    minutes_washing = random.randint(12,18)
    minutes_breakfast = random.randint(5,15)
    minutes_dressing = random.randint(2,8)
    minutes_waiting_bus = random.randint(1,5)
    minutes_bus = random.randint(35,45)
    minutes_on_foot = random.randint(2,8)
    minutes_locker_room = random.randint(4,10)
    minutes = minutes_exercise + minutes_washing + minutes_breakfast + minutes_dressing + minutes_waiting_bus + minutes_bus + minutes_on_foot + minutes_locker_room
    sum_minutes = minutes_exercise + minutes_washing + minutes_breakfast + minutes_dressing + minutes_waiting_bus + minutes_bus + minutes_on_foot + minutes_locker_room
    time_in_minutes.append(sum_minutes)
```

```python
k = int(N * 0.95)
time_curr = time_in_minutes[k]
```

```python
time_of_studying = 12*60 + 50
time_of_waking_up = time_of_studying - time_curr
time_of_waking_up_minutes = time_of_waking_up % 60
if time_of_waking_up_minutes < 10:
    print(f"{time_of_waking_up // 60}:0{time_of_waking_up % 60}")
else:
    print(time_of_waking_up // 60, ":", time_of_waking_up % 60)
```
