```python
import random
import numpy as np
N  = 0
def get_points(N):
    points = []
    for i in range(N):
        x = random.random()
        y = random.random()
        points.append((x,y))
    return points
```

```python
def get_points_in_circle(points):
    points_in_circle = []
    for i in range(len(points)):
        if (points[i][0] - 0.5)**2 + (points[i][1] - 0.5)**2 <= 0.5**2:
            points_in_circle.append(points[i])
    return points_in_circle
points = [get_points(10**2), get_points(10**3), get_points(10**4), get_points(10**5), get_points(10**6), get_points(10**7)]
n = []
for i in range(len(points)):
    n.append(len(get_points_in_circle(points[i])))
    
def pi(n, N):
    return 4*n/N

pi_n = [pi(n[0], 10**2), pi(n[1], 10**3), pi(n[2], 10**4), pi(n[3], 10**5), pi(n[4], 10**6), pi(n[5], 10**7)]
```

```python
import matplotlib.pyplot as plt
from math import log, pi
# ось x - log(N), ось y - pi
x = [log(10**2), log(10**3), log(10**4), log(10**5), log(10**6), log(10**7)]
y = pi_n
plt.plot(x, y)
plt.xlabel('log(N)')
plt.ylabel('pi')
plt.axhline(y=pi, color='green', linestyle='-')
plt.show()
```
